# Activation Functions from Scratch

```python
# --- Core ---
import torch
import matplotlib.pyplot as plt

# --- Display Math Equations in Jupyter ---
from IPython.display import display, Math
```

# Building Activation Functions in PyTorch from Scratch

```python
def plot_activation(func, x_range=(-5, 5)):
    x = torch.linspace(*x_range, 100)
    y = func(x)
    plt.plot(x.numpy(), y.detach().numpy())
    plt.title(func.__name__)
    plt.grid(True)
    plt.show()
```

## ReLU

```python
def relu(x):
    return x * (x > 0)
```

```python
plot_activation(relu)
```

```python
display(Math(fr'ReLU: \ f(x) \ = \ max(0,x)'))
```

## Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))
```

```python
display(Math(r'Sigmoid: \ \sigma (x) \ = \ \frac{1}{1 + e^{-x}}'))
```

```python
plot_activation(sigmoid)
```

## Tanh

```python
def tanh(x):
    return (torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x))
```

```python
display(Math(r'Tanh: \ tanh (x) \ = \ \frac{ e^x - e^{-x} }{ e^x + e^{-x} }'))
```

```python
plot_activation(tanh)
```

## Leaky ReLU

```python
def leaky_relu(x, negative_slope=0.01):
    return torch.where(x > 0, x, x * negative_slope)
```

```python
display(Math(r'\text{Leaky ReLU: } \ f(x) \ = \ max(0.01x, x)'))
```

```python
plot_activation(leaky_relu)
```

## ELU

```python
def elu(x, alpha=1.0):
    return torch.where(x >= 0, x, alpha * (torch.exp(x) - 1))
```

```python
display(Math(r'''
f(x) =
\begin{cases}
x & \text{if } x > 0 \\
\alpha (e^x - 1) & \text{if } x \leq 0
\end{cases}
'''))
```

```python
plot_activation(elu)
```
